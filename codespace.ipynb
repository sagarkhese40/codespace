```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.Figure(figsize=(10,6))

t =np.linspace(0,30 * np.pi,5000)
fs =[1,(1 +np.sqrt(5))/2, np.sqrt(2)]
 
x = np.sin(fs[0]*t) +np.sin(fs[1] *t)
y = np.sin(fs[0]*t) +np.sin(fs[2] *t)

plt.plot(x,y ,color ="grey")
plt.axis('off');
```
